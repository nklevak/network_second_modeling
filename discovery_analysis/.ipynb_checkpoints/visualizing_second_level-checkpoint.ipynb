Code to visualize existing second level model maps (1 per sessio, across subj) for various contrasts.

```python
import os
import numpy as np
import pandas as pd
import nibabel as nib
import nilearn
from nilearn import plotting
from nilearn.glm.first_level import FirstLevelModel
from nilearn.glm.second_level import SecondLevelModel
from nilearn.glm import threshold_stats_img
from nilearn.image import concat_imgs, mean_img, index_img
from nilearn.reporting import make_glm_report
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
# where the second level contrast map pngs are saved:
BASE_DIR_IMGS = '/oak/stanford/groups/russpold/data/network_grant/discovery_BIDS_20250402/derivatives/output_lev2_mni'
CONTRAST_NAMES = ["nBack_twoBack-oneBack", "nBack_task-baseline", "nBack_match-mismatch"] #contrasts of interest
SESSIONS = ['ses-01', 'ses-02', 'ses-03', 'ses-04', 'ses-05', 'ses-06', 'ses-07', 'ses-08', 'ses-09','ses-10']
ENCOUNTERS = ["encounter-1", "encounter-2","encounter-3","encounter-4","encounter-5"]
THRESHOLDS  = ["threshold-1.00.png", "threshold-1.96.png"]
```

```python
# visualize all of the 5 sessions of the nBack_twoBack-oneBack task
contrast_of_interest = CONTRAST_NAMES[0]
threshold_of_interest = THRESHOLDS[1]
basic_filename_twoBackOneBack = os.path.join(BASE_DIR_IMGS, contrast_of_interest)

fig, axes = plt.subplots(5, 1, figsize=(8, 20))  # 5 rows, 1 column
fig.suptitle(f'{contrast_of_interest} across all sessions', fontsize=16)
for idx, encounter in enumerate(ENCOUNTERS):    
    filename = os.path.join(basic_filename_twoBackOneBack, f"{contrast_of_interest}_{encounter}_{threshold_of_interest}")
    img_con = img.imread(filename)

    axes[idx].imshow(img_con)
    axes[idx].set_title(f"Session {idx+1}")
    axes[idx].axis('off') 
    # plt.imshow(img_con)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```
